## 3. ***Data Wrangling***

### Data Wrangling Code

* **Cleaning Rating Column**

````python
# Create a list of ratings from the 'Rating' column
Rating_list=[rating for rating in playstore_df['Rating']]

# Create an empty list to store the cleaned ratings
Data=[]

# Define a function to clean the ratings
def clean_rating(lst):
    # Loop through each rating in the list
    for i in lst:
        # If the rating is greater than 5, set it to 5
        if i > 5:
            i = 5
            # Append the cleaned rating to the Data list
            Data.append(i)
        else:
            # If the rating is within the range of 0 to 5, append it to the Data list
            Data.append(i)
    # Return the cleaned data list
    return Data
````

````python
# Call the clean_rating function on the 'Rating_list' and assign the cleaned ratings to the 'Rating' column of the 'playstore_df' dataframe
playstore_df['Rating'] = clean_rating(Rating_list)

# Convert the 'Rating' column to a float datatype
playstore_df['Rating'] = playstore_df['Rating'].astype(float)
````

````python
#Ploting the histogram to see results
sns.histplot(playstore_df['Rating'])
````

* **Cleaning the size column**

````python
# Create a list of sizes from the 'Size' column
size_list = [size for size in playstore_df['Size']]

# Create an empty list to store cleaned size values
Data=[]

# Create a function to clean the size values
def clean_size(lst):
    for i in lst:
        # Check if 'M' is in the size value
        if 'M' in i:
            # Remove the string 'M' from the size value
            i = i.replace('M', '')
            # Append the cleaned size value to the Data list
            Data.append(i)
        # Check if 'k' is in the size value
        elif 'k' in i:
            # Remove the string 'k' from the size value
            i = i.replace('k', '')
            # Convert size from kb to mb and append to the Data list
            i = float(i) / 1024
            Data.append(i)
        else:
            # Append 0 to the Data list if size value is not in either 'M' or 'k' format
            Data.append(0)
    return Data 
````

````python
# Call the clean_size function on the 'size_list' and assign the cleaned sizes to the 'Size' column of the 'playstore_df' dataframe
playstore_df['Size'] = clean_size(size_list)

# Convert the 'Size' column to a float datatype
playstore_df['Size'] = playstore_df['Size'].astype(float)
````

* **Cleaning Install Column**

````python
# Cleaning the Installs column
Installs_list = [installs for installs in playstore_df['Installs']]
Data = []

def clean_installs(lst):
    """
    This function cleans the installs column by removing '+' and ',' signs and converts the values to integers"""

    for i in lst:
        if '+' in i:
            i = i.replace('+','') # removing the sign '+'
            i = i.replace(',','') # removing the comma ','
            Data.append(i)
        else:
            Data.append(0) # assigning 0 to rows containing 'Free' or 'NaN'
    
    return Data
````

````python
# Call the clean_Install function to clean the Install column
playstore_df['Installs'] = clean_installs(Installs_list)
playstore_df['Installs'] = playstore_df['Installs'].astype(float)
````

* **Cleaning Price Column**

````python
# Create a list of prices from the Price column
price_list = [price for price in playstore_df['Price']]

# Initialize an empty list to store the cleaned data
correct_data = []

# Define a function to clean the price data
def clean_price(lst):
    for i in lst:
        if '$' in i:
            i = i.replace('$', '') # Remove the dollar sign
            correct_data.append(i)
        elif 'Everyone' in i:
            correct_data.append(0) # Replace 'Everyone' with 0
        else:
            correct_data.append(i)
    return correct_data
````

````python
# Call the clean_price function to clean the Price column
playstore_df['Price'] = clean_price(price_list)
playstore_df['Price'] = playstore_df['Price'].astype(float)
````

* **Cleaning Type Column**

````python
#change all the '0' values to 'Free' in the 'Type' column of the 'playstore_df' dataframe
playstore_df['Type'] = playstore_df['Type'].replace('0', 'Free')
````

* **Cleaning Reviews Column.**

````python
# Create a list of reviews by iterating over the 'Reviews' column in the playstore_df dataframe
review_list = [review for review in playstore_df['Reviews']]

# Define a function to clean the reviews
Data = []
def cleaned_review(lst):
  for i in lst:
    if 'M' in i: # Check if the review contains the string 'M' indicating millions
      i = i.replace('M','') # Remove the 'M' string from the review
      i = float(i)*1000000 # Multiply the review by a million to convert to a float value
      Data.append(i)
    else:
      Data.append(i) # If the review does not contain 'M', just append it as is

  return Data
````

````python
# Clean the reviews in the review_list using the cleaned_review function
review_list = cleaned_review(review_list)

# Replace the original 'Reviews' column in the playstore_df dataframe with the cleaned reviews
playstore_df['Reviews'] = review_list

# Change the data type of the 'Reviews' column to float
playstore_df['Reviews'] = playstore_df['Reviews'].astype(float)
````
